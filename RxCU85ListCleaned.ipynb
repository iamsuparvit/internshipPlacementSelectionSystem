```python
import numpy as np
import pandas as pd
import os
import hashlib
```

```python
input_file = './rxcu85_list.csv'
df = pd.read_csv(input_file)
df
```

```python
df.columns = ['student_id','student_pronoun','student_firstname','student_surname','branch','student_nickname']
df
```

```python
df['student_firstname'][1]
```

```python
df['student_name'] = None
```

```python
for i in range(len(df)):
    df['student_name'][i] = f"{df['student_firstname'][i]} {df['student_surname'][i]}"
df
```

```python
df['student_pronoun'].unique().tolist()
```

```python
df['sex'] = df['student_pronoun'].apply(
    lambda x: 'male' if x == 'นาย' else 'female'
)
```

```python
df['shift'] = np.random.choice([1, 2], size=len(df))
```

```python
df_cleaned = df[['student_name','student_id','sex','shift']]
df_cleaned
```

```python
def hash_name(name):
    hashed_object = hashlib.sha256(name.encode('utf-8'))
    return hashed_object.hexdigest()
for i in range(len(df)):
   df_cleaned['student_name'][i] = hash_name(df_cleaned['student_name'][i])
df_cleaned
```

```python
output_path = './rxcu85_list_cleaned_hashed.csv'
df_cleaned.to_csv(output_path, index=False)
```
